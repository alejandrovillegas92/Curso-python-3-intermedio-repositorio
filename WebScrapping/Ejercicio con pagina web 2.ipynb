```python
import requests

from bs4 import BeautifulSoup
```

```python
pagina_web = requests.get('https://www.gonvill.com.mx/libros-de/textos-escolares-y-universitarios-20/?pagSel=1&cuantos=20&orden=prioridad%2C+fecha_edicion+desc&codMateria=20&tipoArticulo=L')


soup = BeautifulSoup(pagina_web.text, 'html.parser')


print(soup.prettify())
```

```python
def obtener_contenido(pagina_web):
    reponse = requests.get(pagina_web)
    return reponse.content
```

```python
def analizar_contenido(pagina_web):
    return BeautifulSoup(pagina_web, 'html.parser')
```

```python
data = []

def procesar_pagina(soup, data):
    titulos = []
    autores = []
    precios = []

    titulo_items = soup.find_all('dd', class_='title')
    for item in titulo_items:
        titulo = item.text.strip()
        titulos.append(titulo)
    

    autor_items = soup.find_all('dd', class_='creator')
    for item in autor_items:
        autor = item.text.strip()
        autores.append(autor)

    precio_items = soup.find_all('p', class_='precio')
    for item in precio_items:
        precio = item.text.strip()
        precios.append(precio)
    
    for i in range(len(titulos)):
        data.append({
            'Titulo':titulos[i] if i < len(titulos) else None,
            'Autor':autores[i] if i < len(autores) else None,
            'Precio':precios[i] if i < len(precios) else None
            })


        
```

```python
url_base = 'https://www.gonvill.com.mx/libros-de/textos-escolares-y-universitarios-20/?pagSel=1&cuantos=20&orden=prioridad%2C+fecha_edicion+desc&codMateria=20&tipoArticulo=L'

num_paginas = 520
```

```python
def paginacion(url_base, num_paginas):
    for i in range(1, num_paginas + 1):
        url = url_base + '&PagSel=' + str(i)
        contenido_pagina = obtener_contenido(url)
        soup = analizar_contenido(contenido_pagina)
        procesar_pagina(soup, data)

paginacion(url_base, num_paginas)
```
