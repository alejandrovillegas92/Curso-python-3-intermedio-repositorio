```python
import requests

from bs4 import BeautifulSoup
```

```python
pagina_web = requests.get('https://listado.mercadolibre.com.mx/celulares-telefonia/celulares-smartphones/xiaomi_Discount_5-100_NoIndex_True#deal_print_id=0bd61b90-2c24-11ee-af00-ebe88f26e1be&c_id=carousel&c_element_order=3&c_campaign=XIAOMI&c_uid=0bd61b90-2c24-11ee-af00-ebe88f26e1be')


soup = BeautifulSoup(pagina_web.text, 'html.parser')


print(soup.prettify())
```

```python
def obtener_contenido(pagina_web):
    reponse = requests.get(pagina_web)
    return reponse.content
```

```python
def analizar_contenido(pagina_web):
    return BeautifulSoup(pagina_web, 'html.parser')
```

```python
data = []

def procesar_pagina(soup, data):
    productos = []
    precios = []
    acumulado_estrellas = []
    existencias = []
    links = []

    producto_items = soup.find_all('h2', class_='ui-search-item__title shops__item-title')
    for item in producto_items:
        producto = item.text.strip()
        productos.append(producto)
        
    precio_items = soup.find_all('span', class_='andes-money-amount__fraction')
    for item in precio_items:
        precio = item.text.strip()
        precios.append(precio)
    
    monto_estrellas_items = soup.find_all('span', class_='ui-search-reviews__amount')
    for item in monto_estrellas_items:
        monto = item.text.strip()
        acumulado_estrellas.append(monto)
    
    existencia_items = soup.find_all('span', class_='ui-search-item__group__element ui-search-item__variations-text shops__items-group-details')
    for item in existencia_items:
        existencia = item.text.strip()
        existencias.append(existencia)
    
    link_items = soup.find_all('a', class_='ui-search-item__group__element shops__items-group-details ui-search-link')
    for item in link_items:
        links.append(item['href'])
    
    for i in range(len(productos)):
        data.append({
            'Producto':productos[i] if i < len(productos) else None,
            'Precio':precios[i] if i < len(precios) else None,
            'Monto acumulado de estrellas':acumulado_estrellas[i] if i < len(acumulado_estrellas) else None,
            'Existencia':existencias[i] if i < len(existencia) else None,
            'Enlaces':links[i] if i < len(links) else None

        })
    
```

```python
url_base = 'https://listado.mercadolibre.com.mx/celulares-telefonia/celulares-smartphones/xiaomi_Discount_5-100_NoIndex_True#deal_print_id=0bd61b90-2c24-11ee-af00-ebe88f26e1be&c_id=carousel&c_element_order=3&c_campaign=XIAOMI&c_uid=0bd61b90-2c24-11ee-af00-ebe88f26e1be' #Deben ser 1088 productos

num_paginas = 1
```

```python
def paginacion(url_base, num_paginas):
    for i in range(1, num_paginas + 1):
        url = url_base +str(i) + '/'
        contenido_pagina = obtener_contenido(url)
        soup = analizar_contenido(contenido_pagina)
        procesar_pagina(soup, data)

paginacion(url_base, num_paginas)
```
